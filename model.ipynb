```python
with open('./train_set.txt', 'r') as f:
    string = f.read()
    
string_refine = string.replace('TrainArticleID~Headline~Body~CompanyID\n0~"', 'TrainArticleID~Headline~Body~CompanyID~0~"')

import re

p = re.compile('[0-9]*?\n[0-9]*?~')

def endlToWave(matched):
    return re.sub('\n', '~', matched.group())

string_refine = re.sub(p, endlToWave, string_refine)
string_refine_split = string_refine.split('~')

pre_table = []

for i in range(int(len(string_refine_split)/4)):
    temp_dict = {
        'AriticleID' : string_refine_split[4*i],
        'Headline' : string_refine_split[4*i+1],
        'Body' : string_refine_split[4*i+2],
        'CompanyID' : string_refine_split[4*i+3]
    }
    pre_table.append(temp_dict)
    
import pandas as pd
data = pd.DataFrame(pre_table)
data = data.drop([0])
```
